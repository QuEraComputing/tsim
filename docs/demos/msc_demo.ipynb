```python
import stim
from tsim.circuit import Circuit
```

```python
c = stim.Circuit.from_file("msc_circuits/d=3.stim")
c.without_noise().diagram("timeline-svg")
```

```python
circuit = Circuit.from_stim_program(c, replace_s_with_t=True)
circuit.diagram()
```

```python
sampler = circuit.compile_detector_sampler()
sampler
```

```python
samples = sampler.sample(100)
print(samples)
```
