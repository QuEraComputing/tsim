```python
from tsim import Circuit
import numpy as np
from numpy.random import rand
from tsim.paulistrings import PauliStrings
```

```python
n = 16

rs = []
for i in range(int(np.log2(n))):
    r = np.arange(n).reshape(-1, 2**i).T.reshape(-1)
    rs.append(" ".join(map(str, r)))



c = Circuit(
    f"""
    R_X({rand()}) 0 
    R_X({rand()}) 3 
    R_X({rand()}) 5 
    R_X({rand()}) 6
    R_X({rand()}) 7 
    R_X({rand()}) 9 
    R_X({rand()}) 10
    R_X({rand()}) 11
    R_X({rand()}) 15
    """)
for i in range(len(rs)):
    c.append_from_stim_program_text(
    f"""
    TICK
    CZ {rs[i]}
    TICK
    SQRT_X_DAG {rs[0]}
    """)

c.diagram(height=400)
```

```python
ps = PauliStrings(c)
```

```python
zeros = np.zeros((n,n), dtype=np.uint8)
eye = np.eye(n, dtype=np.uint8)

x_paulis = np.hstack((eye, zeros))
z_paulis = np.hstack((zeros, eye))

single_paulis = np.vstack((x_paulis, z_paulis))
len(single_paulis)
```

```python
ps.evaluate(single_paulis)
```
